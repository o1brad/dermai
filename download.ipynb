### Downloading the images

```python
from fastai import *
```

```python
import urllib.request

def download_img(url, path):
    urllib.request.urlretrieve(url, path)
    return path


cwd = os.getcwd()
base_path = Path(cwd + '/data/')

for label, urls in labeled_urls:
    n = 0
    for url in urls:
        try:
            download_img(url, cwd + '/data/' + label  + '-' + str(n) + '.jpg')
        except:
            print("invalid url: [}".format(url))
        n += 1
```
